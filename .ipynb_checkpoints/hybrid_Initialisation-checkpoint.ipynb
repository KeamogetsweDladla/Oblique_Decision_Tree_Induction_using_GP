```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler



def initialPopulation(X, frac, n):
    """
    Returns genetic programs based on the 
    
    Parameters:
    X (numpy.ndarray): Feature matrix.
    n (float) population size, e.g 0.6 to represent .
    frac (float)  a percentage of the initial population we aim to initialise.
    
    Returns:
    tuple: Genetic Program Expression representing stacked plane equations  w^T * x + b = 0.
    """
    df = pd.DataFrame(X)

    # Standardize the data
    scaler = StandardScaler()
    scaled_data = scaler.fit_transform(df)

    # Although decision nodes can also split into multiple branches based on the 
    # number of categories in the training data, splitting into two branches is pertinent 
    # in the context of oblique decision trees due to the inequality nature of the expression of the splitting hyperplanes.
    # Therefore, k = 2
    kmeans = KMeans(n_clusters=2)
    kmeans.fit(scaled_data)

    # Get the centroid locations
    centroids = kmeans.cluster_centers_

    # Define the two centroid points
    point1 = centroids[0]
    point2 = centroids[1]
    
    # Calculate the slope of the line connecting the two points
    slope = (point2[1] - point1[1]) / (point2[0] - point1[0])
    
    # Calculate the slope of the perpendicular line
    perpendicular_slope = -1 / slope
    
    # Generate evenly spaced points along the original line
    num_points = math.floor(frac*n)
    x_values = np.linspace(point1[0], point2[0], num_points)
    y_values = np.linspace(point1[1], point2[1], num_points)

    # Calculate the threshold values of the lines
    threshold_values = y_values - perpendicular_slope*x_values

    # Determine the plane tuples
    ones = np.ones(len(y_values))
    x_coeffs = (ones*perpendicular_slope).tolist()
    y_coeffs = ones.tolist()
    planes = tuple[list(zip(x_coeffs,y_coeffs)), list(threshold_values)]    # The planes are in the format [a_1, a_2], (a_0)
    
    # Plot the original line
    plt.plot([point1[0], point2[0]], [point1[1], point2[1]], 'bo-')
    
    # Plot the perpendicular lines
    length = 1  # Length of the perpendicular lines
    for x, y in zip(x_values, y_values):
        dx = length / np.sqrt(1 + perpendicular_slope**2)
        dy = perpendicular_slope * dx
        plt.plot([x - dx, x + dx], [y - dy, y + dy], 'r-')
    
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.title('Perpendicular Lines')
    plt.grid(True)
    plt.show()

    return planes

# Example usage
data = {'feature 1': [1, 2, 8, 4, 2, 1, 9, 3, 2, 7],
        'feature 2': [11, 9, 9, 5, 7, 6, 5, 4, 3, 2]}

planes = initialPopulation(data, 0.10, 100)
print("Planes", planes)
    
```
